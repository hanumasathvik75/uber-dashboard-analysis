```python
import pandas as pd
df = pd.read_csv(r"C:\Users\SATHVIK\OneDrive\Desktop\ncr_ride_bookings.csv")
df.head()
```

```python
df.info()
```

```python
df["Booking ID"]=df["Booking ID"].str.replace('"','')
df["Customer ID"]=df["Customer ID"].str.replace('"','')
```

```python
df["Booking ID"].head()
```

```python
#convert date,time to date time
df["date_time"]=pd.to_datetime(df["Date"]+" "+df["Time"])
```

```python
df=df.drop(columns=["Date","Time"])
```

```python
df.columns
```

```python
#converting null strings to nan
df.replace("null",pd.NA,inplace=True)
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
df["Booking ID"].nunique()
```

```python
len(df)
```

```python
#here duplicate items exists so we insopect them first
df[df["Booking ID"].duplicated(keep=False)].sort_values("Booking ID")
```

```python
#this has an big issue...we should only keep recent records 
df=df.sort_values("date_time")
df=df.drop_duplicates(subset="Booking ID",keep="last")
```

```python
df.duplicated().sum()
```

```python
cols = ["Avg VTAT","Avg CTAT","Booking Value","Ride Distance","Driver Ratings","Customer Rating"]

for col in cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df.info()
```

```python
#we use and create other columns
df["hour"]=df["date_time"].dt.hour
df["month"]=df["date_time"].dt.month
df["day_name"]=df["date_time"].dt.day_name
```

```python
df.columns
```

```python
#
df["Booking Status"].value_counts()
```

```python
#just chercking that whether cancelee rides can have payment or not
df[(df["Booking Status"]=="Cancelled") & (df["Payment Method"].notna())]
```

```python
#replacement null values of payment with 0
df["Payment Method"]=df["Payment Method"].fillna("no-payment")
```

```python
df["Payment Method"].value_counts()
```

```python
df["Payment Method"]=df["Payment Method"].replace(0,"no payment")
```

```python
df["Payment Method"].value_counts()
```

```python
#max of driver and customer rating
print(df["Driver Ratings"].max())
```

```python
print(df["Customer Rating"].max())
```

```python
df["Cancelled Rides by Customer"] = df["Cancelled Rides by Customer"].fillna(0)
df["Cancelled Rides by Driver"] = df["Cancelled Rides by Driver"].fillna(0)
df["Incomplete Rides"] = df["Incomplete Rides"].fillna(0)
```

```python
df["Cancelled Rides by Customer"].value_counts()
```

```python
df.loc[df["Booking Status"] != "Completed", "Ride Distance"] = 0
df.loc[df["Booking Status"] != "Completed", "Booking Value"] = 0
```
